```python
players = len(purchase_df["SN"])

total_players =pd.DataFrame({"Total Players": [len(purchase_df["SN"].unique())]})
total_players
```

```python
Num_Unique_items =len(purchase_df["Item ID"])
avg_price = len(purchase_df["Price"])
total_purchases= len(purchase_df["Purchase ID"])
Total_revenue = sum(purchase_df["Price"])



ending_df = pd.DataFrame({ "Unique Items":Num_Unique_items, "Average Price": "${:.2f}".format(avg_price), "Number of Purchases": total_purchases, "Total Revenue": "${:.2f} ".format(Total_revenue)},index=[0])
ending_df
```

```python
gender_DF = purchase_df.groupby("Gender")
gender_count = gender_DF["Gender"].count()
gender_count



avg_gender = gender_DF["Price"].mean()
avg_gender



gender_total = gender_DF["Price"].sum()
gender_total


avg_per_person = gender_DF["SN"].count()

gender_total_df = pd.DataFrame({"Purchase Count": gender_count, "Average Purchase Price": avg_gender.map("${:.2f}".format), 
                                "Total Purchase Value": gender_total.map("${:.2f}".format), "Avg Total Purchase per Person": avg_per_person.map("${:.2f}".format),})
gender_total_df
```

```python
# age_bins = {"Age": [10, 15, 20, 25, 32, 35, 38, 41, 50]}
# df = pd.DataFrame(age_bins )
# print(df)

bins = [0,9, 14, 19,24,29, 34, 39, 200]
group_names =["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
purchase_df['Age Ranges']= pd.cut(purchase_df["Age"], bins, labels=group_names)
```

```python
players_percentages = purchase_df.groupby('Age Ranges').SN.nunique()

#create dataframe to show total count and percentage of total players for each group
age_demographics = pd.DataFrame({"Total Count": players_percentages, "Percentage of Players": (players_percentages / purchase_df.SN.nunique() * 100).map("{:.2f}".format)
})
age_demographics
```

```python
purchase_data_age = purchase_df[["Age Ranges", "Purchase ID"]]
values_age_df =purchase_df[["Age Ranges", "Price"]]

total_age= purchase_data_age.groupby("Age Ranges").count()
total_values =values_age_df.groupby("Age Ranges").sum()
total_values_person = purchase_data_age["Age Ranges"].count()




Final_df = pd.DataFrame({"Purchase Count" :total_age["Purchase ID"],
                        "Average Purchase Price" :(total_values["Price"]/ total_age["Purchase ID"]).map("${:.2f}".format),
                        "Total Purchase Value":total_values["Price"].map("${:.2f}".format)})
Final_df
```

```python
#Count purchases made by unique SN
spender_purchase = purchase_df["SN"].value_counts()

#Calculate the average purchase by unique SN
spender_avg_purchase = purchase_df["Price"].groupby(purchase_df["SN"]).mean()

#Calculate the total revenue from each unique SN
spender_total_purchase = purchase_df["Price"].groupby(purchase_df["SN"]).sum()


big_spenders = pd.DataFrame({
    "Purchase Count":spender_purchase,
    "Average Purchase Price":spender_avg_purchase,
    "Total Purchase Value":spender_total_purchase
})


big_spenders_sorted = big_spenders.sort_values("Total Purchase Value", ascending= False)
big_spenders_sorted["Average Purchase Price"] = big_spenders_sorted["Average Purchase Price"].map("${:.2f}".format)
big_spenders_sorted["Total Purchase Value"] = big_spenders_sorted["Total Purchase Value"].map("${:.2f}".format)

big_spenders_sorted[:5]
```

```python
pop_items = purchase_df[['Item ID', 'Item Name', 'Price']]


#count transactions for each item
items_purchase = pop_items.groupby(["Item ID","Item Name"])["Price"].count()
#Price of each item
item_price= pop_items.groupby(["Item ID","Item Name"])["Price"].mean()
#Calculate total revenue from each item
popular_purchase_value = pop_items.groupby(["Item ID","Item Name"])["Price"].sum()


pop_items = pd.DataFrame({ "Purchase Count":items_purchase, "Item Price":item_price, "Total Purchase Value":popular_purchase_value})

pop_items_sorted = pop_items.sort_values("Purchase Count", ascending= False)
pop_items_sorted["Item Price"]= pop_items_sorted["Item Price"].map("${:.2f}".format)
pop_items_sorted["Total Purchase Value"]= pop_items_sorted["Total Purchase Value"].map("${:.2f}".format)

pop_items_sorted[:5]
```

```python
#re-sort above table by total purchase
popular_items_purchase = pop_items.sort_values("Total Purchase Value", ascending= False)
popular_items_purchase["Item Price"]= popular_items_purchase["Item Price"].map("${:.2f}".format)
popular_items_purchase["Total Purchase Value"]= popular_items_purchase["Total Purchase Value"].map("${:.2f}".format)

popular_items_purchase[:5]
```
